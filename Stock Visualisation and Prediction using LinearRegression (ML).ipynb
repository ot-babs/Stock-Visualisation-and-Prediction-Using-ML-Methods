<h1 style="text-align:Left; font-size: 1.4em;">Daily Percentage Gain & BoxPlot Visualisation</h1> 

Next, we want to find out the daily percentage gain for each of the stocks. This will be used to help us find out the most volitile stocks, and the best way to visualise this is using a boxplot.

The daily percentage gain is calulated using the following formula: r_t = (pt/(p_(t-1))) - 1


```python
cci['returns'] = (cci['close']/cci['close'].shift(1))-1
amt['returns'] = (amt['close']/amt['close'].shift(1))-1
aapl['returns'] = (aapl['close']/aapl['close'].shift(1))-1
msft['returns'] = (msft['close']/msft['close'].shift(1))-1
gm['returns'] = (gm['close']/gm['close'].shift(1))-1
f['returns'] = (f['close']/f['close'].shift(1))-1
```

<h1 style="text-align:Left; font-size: 2em;">Linear Regression Modelling for Apple Stocks</h1> 

<h1 style="text-align:Left; font-size: 1.4em;">Limitations and Assumptions</h1> 

Our first assumption is that the variables in the data are independent. This dictates that the residuals for any single variable are not related, in this case, the difference between the predicted value and observed value. One issue we have is that our stock values are values of the same thing recorded in sequence since this is Time Series data. This is problematic because it produces a data characteristic that is called autocorrelation

<h1 style="text-align:Left; font-size: 1.4em;">Initialization</h1> 

We will be focusing on Apple stocks because of how volatile the stock has been over the years. First, we pull the data for the apple stocks and then produce a plot of the data.

```python
#Stock Prediction 
#Step one: Keep only the closing values for Apple
aapl_stock = prices.query("symbol == 'AAPL'")

#Plot Closing Prices of the Apple Stocks
aapl_stock['close'].plot()
plt.ylabel('Stock Price ($)')
```

<h1 style="text-align:Left; font-size: 1.4em;">Filtering the Data for modeling</h1> 

Next, we will filter our data for modeling and begin setting up our data for train data and test data.

```python
#Filter the data to just be apple socks
aapl_close = aapl_stock.filter(['close']).values
print(aapl_close)
```

Standardization of the dataset is an important requirement for our machine learning estimators. Using from sklearn.preprocessing import MinMaxScaler. The code allows us to transform features (in this case the closing price of a stock) by scaling each feature to a given range in our case 0 and 1. Fit_transform is used to scale the training data and learn the scaling parameters of the data.

```python
# Scaling the data for modelling
# Also have the time period for the model to predict
from sklearn.preprocessing  import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
aapl_scaled =  scaler.fit_transform(aapl_close)
period = 60

print(aapl_scaled)
```

<h1 style="text-align:Left; font-size: 1.4em;">Developing the Testing - Training Dataset</h1> 

Machine Learning requires a minimum of two data sets for effectiveness and accuracy. The training data and testing data are prerequisites for the model. We will split our training and testing size into 65/35 respectively. These percentages can be changed depending on how accurate our predicted values are.



```python
#We  now split the model into the training and testing dataset
training_size = int(len(aapl_scaled)*0.65)
test_size = len(aapl_scaled)-training_size
train_data, test_data = aapl_scaled[0:training_size,:], aapl_scaled[training_size-60:,:]

x_train =  []
y_train = []
```

Our data has now separated the Dataframe objects to the nearest whole-number reflective of our 65/35 split (which is 1145 training samples, 617 for the test samples). (Note: we have 60 overlapping data points due to the -60). From our split, we obtain the training and testing data. The empty cells for x_train and y_train are going to be used to store our target and feature sets. We will import tqdm which just shows a smart progress meter of the number of iterations completed.

```python
# Preparing thetarget and feature sets

from tqdm import tqdm
for i in tqdm(range(period, len(train_data))):
    x_train.append(train_data[i-period:i,0])
    y_train.append(train_data[i, 0])
    
    #to check what is happening
    # if i <= 60:
    #     print(x_train)
    #     print(y_train)
    #     print()
```

```python
len(x_train), len(y_train)
```

```python
# Now need to convert the features and targets from list to numpy arrays

x_train, y_train = np.array(x_train), np.array(y_train)
```

The next step is to reshape the data, this is because the LSTM model requires 3 dimensions

```python
#Now we need ot convert the size into a size tha we can use LSTM model training
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
x_train.shape
```

LSTM stands for Long Short-Term Memory. We use 50 layers and return sequence true because we will use another layer as an input shape. Next LSTM is 50 layers and then the return sequence is now false since we no longer need new layers for our architecture. We add a dense layer of 25 neurons of densely connected neurons network and then 1 for our last dense layer.

```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, LSTM
```

```python
model = Sequential(
    [
        LSTM(50, input_shape=(x_train.shape[1], 1), return_sequences=True),
        LSTM(50, return_sequences=False),
        Dense(25),
        Dense(1)
    ]
)
```

```python
model.summary()
```

```python
# optimizer improves loss function and loss function is used 
# to measure how well the model did on training
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
#Trianing the Model
model.fit(x_train, y_train,
          batch_size=1,
          epochs=4 )
```

The batch_size is the number of training examples and the epochs are the number of iterations when data is passed back and forth through the Neurol network. As with the training dataset we now have to do the same method for our test dataset. Note: The more epochs the higher the chances of overfititng and the longer the process takes. You could also get a reallu good fit with just 1 epoch.

```python
x_test, y_test = [], aapl_close[training_size:, :]
for i in range(period, len(test_data)):
    x_test.append(test_data[i-60:i, 0])
```

```python
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
len(x_test), len(y_test)
```

Finally, we can get the model to predict stock prices for Apple. Pred is the same as the data in our y_test dataset. Once we have this we convert the predicted dataset back to the original values.

```python
#Get the models predicted price values
pred = model.predict(x_test)

# Cnvert predictd model from its scaled virsion to original values
pred =  scaler.inverse_transform(pred)
```

Next, we want to find the RMSE, coefficient of determination. Our Dataset has an RMSE of 1.9833 which is very accurate. A value of 0 means our model predicted the exact values that we were meant to get from our y_test values.

```python
#Valdating the fit
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

#print("Model Coefficients:", model.coef_)
print("Mean Absolute Error:", mean_absolute_error(y_test, pred))
print("Coefficient of Determination:", r2_score(y_test, pred))
print("Mean Squared Error:", mean_squared_error(y_test, pred, squared = False))
```

<h1 style="text-align:Left; font-size: 2em;">Interpretation of our Analysis</h1> 

We have trained the model on historical pricing data using the closing stock prices and using a 60-day period. The goal was to develop a model that can accurately predict the day’s closing price. Apple stock was chosen because it is a volatile stock.

We will now use our model’s prediction prices for a trading strategy to assess how well our model performs.

<h1 style="text-align:Left; font-size: 1.4em;">Trading Stategy</h1> 

If the model predicts a higher closing stock price than the opening stock price for that day, then we will make a trade on that day. Our budget is $1000 for each day our model tells us to trade the stocks. This means we will buy as the market opens and sell just before the market closes. Then to calculate the gain from the trade we use the difference between the actual market closing price and the open market price.

Assumptions:
- It is possible to buy the stock at the exact open price
- We can sell the stock at the exact closing price


```python
# Asking the model if we should buy or sell

#First convert pandas to array for easy boolean comparison

predictedclose = test['pred'].to_numpy()
openvalue = test['open'].to_numpy()
actualclose = test['close'].to_numpy()

#Boolean for when the trade should be made when the predicted close price is greated than the open value
Trademade = predictedclose > openvalue

# Finding the gain of selling at close price and buying at open price  
gain = []

for k in range(0, len(Trademade)):
    if Trademade[k] == True:
        gain.append(actualclose[k] - openvalue[k])
    else:
        gain.append(0)
```

```python
#add the trademade and gain to our test output
test.insert(loc = 7,
            column='Trade Made',
            value=Trademade)
test.insert(loc=8,
            column='Gain',
            value=gain)
```
